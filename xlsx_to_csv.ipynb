```python
#script to convert all the excel response files into csv 

import pandas as pd
import pathlib 

for path in pathlib.Path("./responses_excel").iterdir():
    df = pd.read_excel(f'{path}')
    csv_name = str(path)[16:-5]
    print(csv_name)
    df.to_csv(f'./RESPONSES/{csv_name}.csv')

    
```
